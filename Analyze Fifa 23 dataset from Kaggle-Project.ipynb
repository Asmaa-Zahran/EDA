```python
def rename(name):
    name = name.lower().strip()
    name = name.replace(' ','_')
    return name
df.columns = map(rename , list(df.columns))
df.columns    
```

```python
df['contract_until'] = df['contract_until'].replace('-',np.nan)
df['contract_until'] = pd.to_datetime(df['contract_until']).dt.year

df['joined_on'] = df['joined_on'].replace('-',np.nan)  
df['joined_on'] = pd.to_datetime(df['joined_on']).dt.year

df['contract_until'] , df['joined_on']
```

```python
# d) check unique values of categorical and their count

for col in cat_cols :
    print(f'count of uniques of \"{col}\" column is :',df[col].nunique(),'\n')
    print(f'uniques of \"{col}\" column is :',df[col].unique())
    print('\n')
    print('=' * 100)
```

## 2) Data Cleaning

* NaNs check and Impute

* Duplicate check and remove

* Outliers check and handle

* Data Inconsistency

```python
 #count nans in each column

countna = df.isna().sum().values

for i in range(len(countna)):
    if countna[i] != 0:
        print(i,df.isna().sum()[i])  
df.isna().sum()[18:20]              
```

```python
df = df.dropna()
df
```

```python
# check duplicates & remove 

df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# delete columns 

deleted_col = list(df.columns[34:])
deleted_col.append('national_team_image_link')
deleted_col.append('image_link')
df.drop(columns=deleted_col,inplace=True)
df
```

```python
# outliers check  

Q1 = df.quantile(.25)
Q1
```

```python
Q3 = df.quantile(.75)
Q3 
```

```python
IQR = Q3 - Q1
IQR
```

```python
num_cols = df.select_dtypes(include='number')
num_cols
```

```python
outliers  = num_cols[(num_cols < Q1 - 1.5 * IQR) | (num_cols > Q3 + 1.5 * IQR)]
print(f"the outliers is : ")
outliers
```

## 3) EDA

**What is the average Age ?**

```python
df['age'].mean()
```

```python
df['age'].median()
```

**What is the greatest Height? and name of the player with that height?**

```python
# greatest height
df['height(in_cm)'].max()
```

```python
# name of the player
df[df['height(in_cm)'] == df['height(in_cm)'].max()]
```

**What is the smallest weight? and name of the player with that weight? and for which club plays?**

```python
df['weight(in_kg)'].min()
```

```python
# name of the player and his club

df[df['weight(in_kg)'] == df['weight(in_kg)'].min()][['known_as', 'club_name']]
```

**What is the position of Cristiano Ronaldo?**

```python
df[df['known_as'] == 'Cristiano Ronaldo']['club_position'].values[0]
```

**What was the average (mean) Value of all players per year(Joined column) ?**

```python
df.groupby('joined_on')[['wage(in_euro)']].mean().sort_values( by='wage(in_euro)',ascending=False)
```

**get Number of players for each club**

```python
df.groupby('club_name')[['known_as']].count()
```

**How many unique Clubs are there?**

```python
df['club_name'].value_counts()
```

**What are the top 5 most common positions?**

```python
df.groupby('club_position')[['known_as']].count().sort_values(by='known_as',ascending=False).head()
```

**What are the top 5 most common nationality?**

```python
df.groupby('nationality')[['known_as']].count().sort_values(by='known_as',ascending=False).head()
```

**What are the names of players has Egyptian Nationality?**

```python
df[df['nationality'] == "Egypt"][['known_as']]
```

**How many players joined in 2018 with skill moves=2?** 

```python
df[(df['joined_on'] == 2018) | (df['skill_moves'] == 2)].shape[0]
```

```python
df['joined_on'] = np.datetime64()
df[(df['joined_on'] >= np.datetime64('2018-01-01') ) & (df['skill_moves'] == 2)].shape[0]
```

**How many players whom name starts with letter  'M'? (This is pretty tricky)**

```python
df[ (df['known_as'].str[0] == 'M') | (df['known_as'].str[0] == 'm') ].shape[0]
```

**Bonus: Is there a correlation between International Reputation and Wage?**

```python
df[['international_reputation','wage(in_euro)']].corr() # 0.63 - > high corr 
```
